## 01-04 Computation Graph

```python
import torch

torch.__version__
```

分析下述式子的计算图  `y=(w+x)(w+1)`

```python
w = torch.tensor([1.], requires_grad=True)
x = torch.tensor([2.], requires_grad=True)

a = torch.add(w, x)
a.retain_grad()  # 保存非叶子节点的梯度
b = torch.add(w, 1)
y = torch.mul(a, b)

print(y)
```

反向传播计算梯度

```python
y.backward()
print(w.grad)
```

查看叶子节点

```python
print(w.is_leaf, x.is_leaf, a.is_leaf, b.is_leaf, y.is_leaf)
```

查看梯度

```python
print(w.grad, x.grad, a.grad, b.grad, y.grad)
```

查看`grad_fn`

```python
print(w.grad_fn, x.grad_fn, a.grad_fn, b.grad_fn, y.grad_fn)
```
